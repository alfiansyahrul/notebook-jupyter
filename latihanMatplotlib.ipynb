```python
import matplotlib.pyplot as plt
```

```python
# siapkan data yang akan di tampilkan
import numpy as np
# Membuatb data array di mulai dari angka 1 sampai 5 dengan jumlah index 11
x = np.linspace(1,5,11)
# dari aaray yang telah di buat di atas di beri pangkat 2
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Membuat plot dari data x dan y
plt.plot(x,y,'r') # r merupakan warna yaitu red
plt.xlabel('X Axis') # label untuk sumbu x
plt.ylabel('Y Axis') # label untuk sumbu y
plt.title('Data Contoh') # labe untuk judul diagram
plt.show() # untuk menampilkan diagram
```

```python
# Membuat banyak diagram atau plot di 1 canvas
plt.subplot(1,2,1) # 1 adalah row 2 adalah kolom 1 adalah plot
plt.plot(x,y,'r--') # r-- adalah garis berwarna putus - putus
plt.subplot(1,2,2) # 1 adalah row 2 adalah kolom 1 adalah plot
plt.plot(y,x,'b*-') # b*- adalah garis baru berwarna biru dengan campuran bintang dan garis
```

```python
# membuat kanvas kosong
fig = plt.figure()
# mengatur ukuran kanvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# 0.1 jarak dari kiri
# 0.1 jarak dari bawah
# 0.8 lebar
# 0.8 tinggi

axes.plot(x,y, 'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Contoh Judul')
```

```python
# plot didalam plot
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # plot utama
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # plot didalam

# pengaturan plot utama
axes1.plot(x,y, 'b')
axes1.set_xlabel('X utama')
axes1.set_ylabel('Y utama')
axes1.set_title('Plot Utama')

#pengaturan plot dalam
axes2.plot(x,y, 'r')
axes2.set_xlabel('X utama')
axes2.set_ylabel('Y utama')
axes2.set_title('Plot Dalam')
```

```python
# Mengatur canvas
# figsize adalah pengaturan utama ukuran terhadap canvas yang aka dibuat
# dpi adalah bentukan dari suatu ukuran hasil cetak atau visual semakin tinggi maka semakin besar gambarnya
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,4))

axes.plot(x,y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# save atau export Figure
fig, axes = plt.subplots(figsize=(8,4))

axes.plot(x,y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');

# save atau export figure menjadi format PNG, JPG, EPS, SVG, PGF, dan PDF
fig.savefig("contoh.png", dpi=200)
```
