```python
import pandas as pd
```

```python
# buat dataframe
data  = {'Perusahaan':['Babastudio','Babastudio','Techfor','Techfor'],
        'Nama':['Nur','Reza','Gusti','Pandu'],
        'Penjualan':[245,250,350,400]}
```

```python
df = pd.DataFrame(data)
df
```

```python
# GroupBy
# Membuat group untuk kolom perusahaan
df.groupby('Perusahaan')
```

```python
# Menampilkan rata-rata penjualan dari masing-masing perusahaan
df.groupby('Perusahaan').mean()
```

```python
# Menampilkan data terkecil dari masing-masing group
df.groupby('Perusahaan').min()
```

```python
# Menampilkan data terbesar dari masing-masing group
df.groupby('Perusahaan').max()
```

```python
# Menampilkan jumjlah record dari masing-masing perusahaan
df.groupby('Perusahaan').count()
```

```python
# Menampilkan data-data dari group
df.groupby('Perusahaan').describe()
```

```python
# Menampilkan data-data group secara horizontal
df.groupby('Perusahaan').describe().transpose()
```

```python
# Menampilkan data-data peusahaan Babastudio
df.groupby('Perusahaan').describe().transpose()['Babastudio']
```

```python
# Concat
df1 = pd.DataFrame({'A':['A0','A1','A2','A3',],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                  index = [0,1,2,3])
df1
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7',],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                  index = [4,5,6,7])
df2
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11',],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                  index = [8,9,10,11])
df3
```

```python
# Concat
pd.concat([df1,df2,df3])
```

```python
# Bisa juga menggunakan axis
pd.concat([df1,df2,df3],axis=1)
```

```python
# Merge
left = pd.DataFrame({'key1': ['K0','K0','K1','K2'],
                    'key2': ['K0','K1','K0','K1'],
                    'A' : ['A0','A1','A2','A3'],
                    'B' : ['B0','B1','B2','B3']})
left
```

```python
right = pd.DataFrame({'key1': ['K0','K1','K1','K2'],
                    'key2': ['K0','K0','K0','K0'],
                    'C' : ['C0','C1','C2','C3'],
                    'D' : ['D0','D1','D2','D3']})
right
```

```python
# How = inner (tipe merger)
# Dimana ketika tidak ada yang cocok antara key1 dan key2  maka dia tidak akan tampil
pd.merge(left,right,on=['key1', 'key2'])
```

```python
# How = outer
# Dimana data akan ditampilkan semua ketika ada data yang tidak cocok maka akan ditulis Nan
pd.merge(left,right, how='outer', on=['key1','key2'])
```

```python
# How = right
# Dimana dataframe right akan ditampilkan semua dan dataframe left akan ditampilkan jika sesuai dengan data right
pd.merge(left,right, how='right', on=['key1','key2'])
```

# Note : 
#1. Right key1 dan key2 index 0 sama dengan left key1 dan key2 index 0 menjadikan A dan B ditampilkan
#2. Right key1 dan key2 index 1 sama dengan left key1 dan key2 index 2 menjadikan A dan B ditampilkan
#3. Right key1 dan key2 index 2 sama dengan left key1 dan key2 index 2 menjadikan A dan B ditampilkan
#4. Right key1 dan key2 index 3 tidak sama dengan left key1 dan key2 di index manapun menjadikan nilai a dan B  menjadi NaN

```python
# How left
# Dimana dataframe left akan ditampilkan seuma dan dataframe right akan ditampilkan jika sesuai dengan dddata left
pd.merge(left,right, how='left', on=['key1','key2'])
```

# Note
#1. Left key1 dan key2 index 0 sama dengan  right key1 dan  key2 index 0 menjadikan C dan D ditampilkan
#2. Left key1 dan key2 index 1 tidak sama dengan right key1 dan key2 di index manapun menjadikan nilai C dan D menjadi NaN
#3. Left key1 dan key2 index 2 sama dengan  right key1 dan  key2 index 1 dan index 2 menjadikan dia ditampilkan 2 data C dan D ditampilkan
#4. Left key1 dan key2 index 3 tidak sama dengan right key1 dan key2 di index manapun menjadikan nilai C dan D menjadi NaN

```python
# JOIN
left = pd.DataFrame({'A':['A0','A1','A2'],
                    'B':['B0','B1','B2']},
                    index=['K0','K1','K2'])
right = pd.DataFrame({'C':['C0','C1','C2'],
                    'D':['D0','D1','D2']},
                    index=['K0','K2','K3'])
```

```python
# How = inner
left.join(right, how='inner')
```

```python
# How = outer
left.join(right, how='outer')
```

```python
# How = left
left.join(right, how='left')
```

```python
# how = right
left.join(right, how='right')
```
